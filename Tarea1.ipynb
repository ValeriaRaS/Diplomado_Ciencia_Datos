## 1.4 ¬øQu√© es el promedio matematico o media?

El promedio es un valor "central" calculado entre un conjunto de n√∫meros. Es un n√∫mero representativo que puede obtenerse a partir de una lista de cifras. Usualmente se relaciona con el concepto de media aritm√©tica.

Es f√°cil de calcular: suma todos los n√∫meros y divide por la cantidad de n√∫meros que hay, y se obtiene el promedio.

 Promedio = $\bar{x}$ = $\frac{x_1 + x_2 + ... + x_n}{ n}$

### Ejemplo: Promedio de 3, 7, 15 y 10

```python
Promedio =  (3+7+15+10)/4
Promedio
```
